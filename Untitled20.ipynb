 Q4) What is a box plot, and what can it tell you about the distribution of data?
 -->A box plot, also known as a box-and-whisker plot, is a graphical representation of a dataset's distribution based on its five-number summary: minimum, first quartile (Q1), median, third quartile (Q3), and maximum. It provides a visual summary of the data's spread, central tendency, and potential outliers.

Here's what a box plot can tell you:

Spread: The box represents the interquartile range (IQR), which contains the middle 50% of the data. The length of the box indicates the spread of the middle portion of the data.
Central Tendency: The line inside the box represents the median, which is the middle value of the dataset.
Skewness: The position of the median within the box can indicate skewness. If the median is closer to the bottom of the box, the data is skewed to the right (positively skewed). If it's closer to the top, the data is skewed to the left (negatively skewed).
Outliers: The whiskers extend from the box to the minimum and maximum values, excluding outliers. Outliers, which are data points that fall outside the typical range, are often plotted as individual points beyond the whiskers.

 Q7) What is the interquartile range (IQR), and how is it used to detect outliers?
 -->The interquartile range (IQR) is a measure of statistical dispersion, or variability, between the upper quartile (Q3) and the lower quartile (Q1) in a data set. It represents the range for the middle 50% of the data.

Calculating IQR:

Order the data: Arrange the data points in ascending order.
Find the median: Determine the median (Q2) of the data set.
Find Q1: Determine the median of the lower half of the data (below Q2).
Find Q3: Determine the median of the upper half of the data (above Q2).
Calculate IQR: Subtract Q1 from Q3: IQR = Q3 - Q1.
Using IQR to Detect Outliers:

Outliers are data points that fall significantly outside the typical range of the data. The IQR can be used to identify potential outliers using the following rule:

Lower Boundary: Q1 - (1.5 * IQR)
Upper Boundary: Q3 + (1.5 * IQR)
Any data point that falls below the lower boundary or above the upper boundary is considered a potential outlier

Q9)Explain the properties of the normal distribution and the empirical rule (68-95-99.7 rule)

-->Properties of the Normal Distribution:

Bell-shaped curve: The distribution is symmetric around the mean, resembling a bell.
Mean, median, and mode are equal: In a perfectly normal distribution, these central tendency measures coincide.
Infinite range: The distribution extends infinitely in both positive and negative directions, although the probability of observing values far from the mean becomes extremely small.
Standard deviation: The standard deviation determines the spread or width of the bell curve. A larger standard deviation indicates a wider distribution.
The Empirical Rule (68-95-99.7 Rule)

The empirical rule, also known as the three-sigma rule, provides a quick way to estimate the proportion of data that falls within specific intervals around the mean in a normal distribution:

68%: Approximately 68% of the data falls within one standard deviation of the mean.
95%: Approximately 95% of the data falls within two standard deviations of the mean.
99.7%: Approximately 99.7% of the data falls within three standard deviations of the mean.

Q10) Provide a real-life example of a Poisson process and calculate the probability for a specific event.

-->A website receives an average of 20 hits per minute. Assuming the hits follow a Poisson process, what is the probability that the website receives exactly 25 hits in the next minute?

Solution:

Define the parameters:

λ (lambda) = average number of hits per minute = 20
k = desired number of hits = 25
Use the Poisson probability formula:

P(X = k) = (e^(-λ) * λ^k) / k!
where:

e is Euler's number (approximately 2.71828)
k! is the factorial of k (k * (k-1) * (k-2) * ... * 1)
Calculate:

P(X = 25) = (e^(-20) * 20^25) / 25!
Using a calculator or statistical software, we find:

P(X = 25) ≈ 0.0446
Interpretation:

The probability that the website receives exactly 25 hits in the next minute is approximately 0.0446 or 4.46%.

Calculate the mean of X and Y:

Mean of X (X̄) = (2 + 5 + 3 + 4 + 1) / 5 = 3
Mean of Y (Ȳ) = (60 + 80 + 70 + 75 + 50) / 5 = 67
Calculate the deviations from the mean for each data point:

For X: (2 - 3), (5 - 3), (3 - 3), (4 - 3), (1 - 3) = -1, 2, 0, 1, -2
For Y: (60 - 67), (80 - 67), (70 - 67), (75 - 67), (50 - 67) = -7, 13, 3, 8, -17

(-1) * (-7) = 7
2 * 13 = 26
0 * 3 = 0

Sum the products of deviations:

7 + 26 + 0 + 8 + 34 = 75
Divide the sum by (n - 1), where n is the number of data points:

Covariance = 75 / (5 - 1) = 75 / 4 = 18.75
Calculating Correlation:

Calculate the standard deviation of X and Y:

Standard deviation of X (σX) ≈ 1.58
Standard deviation of Y (σY) ≈ 11.18
Divide the covariance by the product of standard deviations:

Correlation (ρ) = Covariance / (σX * σY) = 18.75 / (1.58 * 11.18) ≈ 0.998
